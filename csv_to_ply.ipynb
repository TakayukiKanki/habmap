```python
vertice_data=pd.read_csv("vertice_dataset.csv", header=None)
faces_data=pd.read_csv("faces_dataset.csv", header=None)
property_data=pd.read_csv("property_data.csv", header=None)
```

```python
#編集したarrayをplyファイルに直すのに使います
def array_to_ply(property_data, vertice_data, faces_data):
    vertice_data2=[]
    for line in vertice_data:
        line2 = map(str, line)  #mapで要素すべてを文字列に
        vertice_data2.append(' '.join(line2))

    faces_data2=[]
    for line in faces_data:
        line2 = map(str, line)  #mapで要素すべてを文字列に
        faces_data2.append(' '.join(line2))
    plydata=property_data+vertice_data2+faces_data2
    return(plydata)

#plyデータの保存
def save_ply(model_path, plydata):
    with open(model_path, mode='w') as f:
        f.writelines(plydata)
```
